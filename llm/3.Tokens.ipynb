```python
import tiktoken

encoding = tiktoken.encoding_for_model("gpt-5-nano")

tokens = encoding.encode("Hi my name is Yashwanth and I like Pizza")
```

```python
tokens
```

```python
for token_id in tokens:
    token_text = encoding.decode([token_id])
    print(f"{token_id} = {token_text}")
```
